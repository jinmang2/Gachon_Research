```python
import sys 
mod = sys.modules[__name__]

ca_total = ca(sanitation)
for gni in income.keys():
    setattr(mod, 'ca_{}'.format(income[gni]), 
            ca(sanitation[sanitation['IncomeGroup']==gni]))
for re in region.keys():
    setattr(mod, 'ca_{}'.format(region[re]), 
            ca(sanitation[sanitation['region']==re]))

ca_total.t_test()
ca_total.chi_square_test()
for gni in income.keys():
    getattr(mod, 'ca_{}'.format(income[gni])).t_test()
    getattr(mod, 'ca_{}'.format(income[gni])).chi_square_test()
for re in region.keys():
    getattr(mod, 'ca_{}'.format(region[re])).t_test()
    getattr(mod, 'ca_{}'.format(region[re])).chi_square_test()
```
